```python
import numpy as np
import cv2
import os
import time
from cpplib.cpp_flow import Flow
```

## Get Video Images

```python
video_path = 'videos/01'
```

```python
image_ext = ['jpg', 'jpeg', 'png']
```

```python
if os.path.isdir(video_path):
    image_names = []
    ls = os.listdir(video_path)
    for file_name in sorted(ls):
        ext = file_name[file_name.rfind('.')+1:].lower()
        if ext in image_ext:
            image_names.append(os.path.join(video_path, file_name))
else:
    print('Invalid Path')
    assert 0
```

## Optical Flow

```python
prev_img = cv2.imread(image_names[0])
prev_img = cv2.cvtColor(prev_img, cv2.COLOR_BGR2GRAY)
prev_img = cv2.resize(prev_img, (640, 360), interpolation=cv2.INTER_AREA)
cpp_flow = Flow(prev_img)

for image_name in image_names[1:]:
    cur_img = cv2.imread(image_name)
    cur_img = cv2.cvtColor(cur_img, cv2.COLOR_BGR2GRAY)
    cur_img = cv2.resize(cur_img, (640, 360), interpolation=cv2.INTER_AREA)
    
    # dense optical flow
    stime = time.time()
    flow = cpp_flow.get_flow(cur_img)
    etime = time.time()
    print('flowtime = {}'.format(etime-stime))
    
    flow = flow.reshape(360, 640, 2)
    
    # only for visuailization
    flow_horz = np.abs(flow[:,:,0])
    flow_horz /= np.max(flow_horz) 
    flow_vert = np.abs(flow[:,:,1])
    flow_vert /= np.max(flow_vert)
    cv2.imshow('cur_img', cur_img)
    cv2.imshow('flow_horizontal', flow_horz)
    cv2.imshow('flow_vertical', flow_vert)
    
    k = cv2.waitKey(0)
    if k == 27:
        break
    
cv2.destroyAllWindows()
```
